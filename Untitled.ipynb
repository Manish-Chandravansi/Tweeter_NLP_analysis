# Text Analysis

```python
# Importing the Nueral Language ToolKit module
import nltk
```

```python
# Importing more advance function from nltk
import re
nltk.download('punkt')
nltk.download('stopwords')
from nltk.corpus import stopwords
```

# data preprocessing 

```python
# converting pandas text data into list
tweet_text=[]
for i in range(len(tweets)):
    tweet_text.append(str(tweets["text"][i]))
```

```python
tweet_text
```

```python
len(tweet_text)
```

```python
# Remove white spaces
tweet_new=[]
for i in range(len(tweet_text)):
    tweet_new.append(re.sub("\s+"," ", tweet_text[i]))
print(tweet_new)
```

```python
# Now case normalisation into all to lower case
tweet_lower=[]
for i in range(len(tweet_new)):
    tweet_lower.append(tweet_new[i].lower())
```

```python
tweet_lower
```

```python
# here we remove the Http contains since they have no use for analysis
tweet_http=[]
for i in range(len(tweet_lower)):
    tweet_http.append(re.sub('(http\S+)', ' ', tweet_lower[i]))
tweet_http
```

```python
# here we remove the @ contains since they have no use for analysis
tweet_atr=[]
for i in range(len(tweet_http)):
    tweet_atr.append(re.sub("(@[a-z0-9]+)\w+", " " , tweet_http[i]))
tweet_atr
```

```python
# here we remove the hashtag contains since they have no use for analysis
tweet_hass=[]
for i in range(len(tweet_atr)):
    tweet_hass.append(re.sub("(#[a-z0-9]+)\w+", " " , tweet_atr[i]))
tweet_hass
```

```python
len(tweet_hass)
```

```python
# Here we remove the pantuations since they have no use
tweet_panch=[]
for i in range(len(tweet_hass)):
    tweet_panch.append(re.sub("[^0-9a-z \t]", " " , tweet_hass[i]))
tweet_panch
```

```python
# here we are importing tweeter tokenizer 
from nltk.tokenize import TweetTokenizer
```

```python
# converting into the list of list
tweet = TweetTokenizer()
tweet_word=[]
for i in range(len(tweet_panch)):
    tweet_word.append(tweet.tokenize(tweet_panch[i]))
```

```python
tweet_word
```

```python
len(tweet_word)
```

```python
# here we are preparing list for the text length
text_length=[]
for i in range(len(tweet_word)):
    text_length.append(len(tweet_word[i]))
```

```python
text_length
```

```python
# now stop word removal
stop_words=stopwords.words('english')
```

```python
stop_words
```

```python
# Now here we would remove the stop words since they have no use to us
process_word=[]
for i in range(len(tweet_word)):
    process_word.append([x for x in tweet_word[i] if x not in stop_words])
process_word
```

# Sentiment analysis

```python
#count number of positive words
positive_words_count=lambda x: len([w for w in x if w in positive_word])
#count number of negative words
negative_words_count=lambda x: len([w for w in x if w in negative_word])
#count number of offensive words
offensive_words_count=lambda x: len([w for w in x if w in offensive_words])
#count number of abusive words
abuse_vulgar_count=lambda x: len([w for w in x if w in abuse_vulgars])
```

```python
no_positive=[]
no_negative=[]
polarity=[]
no_offensive=[]
no_abuse_vulgar=[]
for i in range(len(process_word)):
    positive_count=positive_words_count(process_word[i])
    no_positive.append(no_positive)
    no_negative.append(negative_words_count(process_word[i]))
    polarity.append(((positive_words_count(process_word[i])-negative_words_count(process_word[i]))/(len(process_word[i]))+0.000001))
    no_offensive.append(offensive_words_count(process_word[i]))
    no_abuse_vulgar.append(abuse_vulgar_count(process_word[i]))
```

```python
# from above its clear that there is process word having zero length so 
# here is the updated process
no_positive=[]
no_negative=[]
polarity=[]
no_offensive=[]
offensiveness=[]
no_abuse_vulgar=[]
abusive_level=[]
for i in range(len(process_word)):
    positive_count=positive_words_count(process_word[i])
    no_positive.append(positive_count)
    no_negative.append(negative_words_count(process_word[i]))
    if len(process_word[i])==0:
        polarity.append(0)
    else:
        polarity.append(((positive_words_count(process_word[i])-negative_words_count(process_word[i]))/(len(process_word[i]))+0.000001))
    no_offensive.append(offensive_words_count(process_word[i]))
    if len(process_word[i])==0:
        offensiveness.append(0)
    else:
        offensiveness.append(offensive_words_count(process_word[i])/text_length[i])
    no_abuse_vulgar.append(abuse_vulgar_count(process_word[i]))
    if len(process_word[i])==0:
        abusive_level.append(0)
    else:
        abusive_level.append(abuse_vulgar_count(process_word[i])/text_length[i])
```

```python
# here polarity count the tendency of word whether it is positive or negative
polarity
```

```python
# assingning sentiment to each tweet
sentiment=[]
for i in range(len(process_word)):
    if polarity[i]>0.1:
        sentiment.append("Positive")
    elif polarity[i]<-0.1 or abusive_level[i]>0:
        sentiment.append("Negative")
    else:
        sentiment.append("Neutral")
```

```python
# creating empty dataframe analysis
analysis=[]
analysis=pd.DataFrame(analysis)
```

```python
# now creating list with above obtain values
analysis["positive_score"]=no_positive
analysis["negative_score"]=no_negative
analysis["polarity"]=polarity
analysis["word_count"]=text_length
analysis["Sentiment"]=sentiment
analysis["offensive_score"]=no_offensive
analysis["offensiveness"]=offensiveness # value 1 is highly offensive
analysis["abusive_volgarity_score"]=no_abuse_vulgar
analysis["abusive_volgarity_level"]=abusive_level # value one is highly abusive
```

```python
analysis
```

```python
# preparing a final list here with each tweet
Tweet_analysis= pd.concat([tweets,analysis], axis=1)
```
